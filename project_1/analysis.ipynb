# Udemy Courses

The data this analysis covers comes from the dataset 'Udemy Courses', available on:

https://www.kaggle.com/datasets/hossaingh/udemy-courses


It consists on a dataset with the main characteristics of online courses in Udemy.

# Problem definition

This notebook aims to analyse the data present in the dataset about online courses on Udemy, to try to measure what makes a course successful.

Before we can ask any question to answer the main objective, we must understand what does the data contain.

# Understanding the data

The dataset contains the following data:

- id: Id of the course
- title: Title of the course
- is_paid: Whether the course is free or paid
- price: Price in USD
- headline: The courses headline
- num_subscribers: How many people bought/subscribed to the course
- avg_rating: Average rating of the people that completed the course
- num_reviews: Number of reviews left by users
- num_comments: Number of comments that the course have
- num_lectures: Amount of lectures the couse has
- content_lenght_min: Total of minutes of all videos present in the course
- last_update_date: time of the last update
- category: type of content the course has
- language: language the course was published
- course_url: link to the course's webpage
- instructor_name: Name of the instructor that created the course
- instructior_url: Link to the instructors page

# Assumptions

During the analysis of this data, the following assumptions will be made:
1. Every subscriber actually watch some part of the course before leaving a review
2. Only people that subscribed are able to leave a rating and comment
3. Every rating was honest

# Analysis

The objective of this analysis is to differentiate the main characteristics between the top earners and the rest of paid curses, to identify what contributes to the success of a course.

# Loading the dataset

Now that we know what we are looking for, lets starting the analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/Course_info.csv')
print(df.shape)
df.head()
```

```python
df.info()
```

# Data preparation

Before we can use the data, first it needs to be checked for inconsistencies and cleaned

```python
df.isna().sum()
```

Of all the columns containing Nan the main concern is the topic data, which will be used in this analysis. Considering the dataset contains 209734 entries, 958 rows correspond to less than 1% of the total, so they will simply be dropped

```python
df = df.dropna(subset=['topic'])
print(df.shape)
df.isna().sum()
```

The other columns with Nan values are not going to be used in this analysis, so they are not a concern.

Other columns tha will be dropped are: title, published_time, last_update_date, language, course_url and instructor_url.
Headline will also be dropped, but first it will be transformed in a boolean value named hasHeadline, to show if the publisher included a headline in the course

```python
df['hasHeadline'] = df['headline'].apply(lambda x: len(str(x)) >= 2)
```

```python
df.head()
```

```python
df[df.hasHeadline == False].head()
```

Only 5 courses in the dataset don´t have a significant headline, so this is unlikely to impact the income

```python
# drop title, headline, published_time, last_update_date, language, course_url and instructor_url
df = df.drop(columns=['title', 'headline', 'hasHeadline', 'published_time', 'last_update_date', 'language', 'course_url', 'instructor_url', 'instructor_name'])
df.head() 
```

The last thing I want to do is to convert the id and num_subscribers columns to int64. This is a small thing, but might prevent errors down the line

```python
df.id = df.id.astype(np.int64)
df.num_subscribers = df.num_subscribers.astype(np.int64)
```

```python
df.head()
```

```python
df.isna().sum()
```

# Exploratory Data Analysis

Now that the data is prepared, lets start answering our questions

```python
df = df.copy()
```

The first difference to be checked will be the difference in income, to see how different it is

```python
df['income'] = df.apply(lambda x: (x.price * x.num_subscribers)/1000, axis=1)
df.head(5)

df = df[df.is_paid == True]
```

```python
# Separate the top 1000 by income in a dedicated df and put the rest in another

top1000 = df.sort_values(by=['income'], ascending=False)[:1000]
others = df.sort_values(by=['income'], ascending=False)[1000:]

others = others[others.income > 1000]
```

```python
# Plot income differences between both groups

plt.violinplot(dataset = [top1000['income'].values,
                           others['income'].values ] )

plt.xticks([1, 2], labels=['top 1000', 'others'])
plt.show()
```

The plot shows a significant difference, where some of the high earners bring more than 300 million usd in income. The top 100 also tend to earn more on average, but in the previous plot it is difficult to see. Lets remove the courses above 20 million to see better the difference

```python
# Since the range is so different, lets plot the income up to 20000k
plt.violinplot(dataset = [top1000[top1000.income <= 20000]['income'].values,
                           others[others.income <= 20000]['income'].values ] )

plt.xticks([1, 2], labels=['top 1000', 'others'])
plt.show()
```

Not only they earn more, but the average is also higher

### Rating

Lets see the difference in rating between the two classes

```python
# Difference in rating
plt.violinplot(dataset = [top1000.avg_rating.values,
                           others.avg_rating.values ] )

plt.xticks([1, 2], labels=['top 1000', 'others'])
plt.show()
```

Also there isn´t much, only that no course in the top 1000 is rated bellow 2.5

### Subscribers

Let´s see the impact of subscribers. A course can earn more by being more expensive, or by having more subscribers

```python
# Difference in number of subscribers
plt.violinplot(dataset = [top1000.num_subscribers.values,
                           others.num_subscribers.values ] )

plt.xticks([1, 2], labels=['top 1000', 'others'])
plt.show()
```

Here it is also difficult to see. Lets remove the courses with more than 600k subscribers

```python
plt.violinplot(dataset = [top1000[top1000.num_subscribers <= 0.6*1e6]['num_subscribers'].values,
                           others[others.num_subscribers <= 0.6*1e6]['num_subscribers'].values ] )

plt.xticks([1, 2], labels=['top 1000', 'others'])
plt.show()
```

The highest earning also tend to have more subscribers

### Price

Let´s see the difference in price between the two classes

```python
# Difference in price
plt.violinplot(dataset = [top1000.price.values,
                           others.price.values ] )

plt.xticks([1, 2], labels=['top 1000', 'others'])
plt.show()
```

It shows that, on average, both classes have the same price.

### Length of content


```python
# Difference in length of the content
plt.violinplot(dataset = [top1000.content_length_min.values,
                           others.content_length_min.values ] )

plt.xticks([1, 2], labels=['top 1000', 'others'])
plt.show()
```

```python
plt.violinplot(dataset = [top1000[top1000.content_length_min <= 4000]['content_length_min'].values,
                           others[others.content_length_min <= 4000]['content_length_min'].values ] )

plt.xticks([1, 2], labels=['top 1000', 'others'])
plt.show()
```

In the first plot it is difficult to see the difference, but if it is limited to a maximum of 4000 minutes a few differences can be seen.
The first is that the courses outside the top 1000 tend to have shorter duration and then drop fast, whereas the top 1000 average around 250 minutes but also have more examples in the 1000 to 2500 minutes

### Categories

Lets see which categories are more represented in each class

```python
plt.figure(figsize=(15, 5))

top1000_cat = top1000.subcategory.groupby(top1000.subcategory).count().sort_values(ascending=True)[-20:]

plt.barh(top1000_cat.index, top1000_cat.values, align='center')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15, 5))

others_cat = others.subcategory.groupby(others.subcategory).count().sort_values(ascending=True)[-20:]

plt.barh(others_cat.index, others_cat.values, align='center')
plt.yticks()
plt.xticks(rotation=90)
plt.show()
```

The top 2 in both cases are web development and programming languages, but their representation are very different. These two represent the majority of courses in the top 1000, but are less dominant in the others. Data science is the third best ranked in the top 1000, but falls to 7th in the rest. 

The other key takeway is the the top earners are mainly in the tech categories.
